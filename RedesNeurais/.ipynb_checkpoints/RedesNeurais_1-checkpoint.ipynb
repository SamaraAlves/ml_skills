```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3d
import tensorflow as tf
```

#### Gerar dados aleat√≥rios 

```python
observations = 1000
variabels = 1

axes_x = np.random.uniform(low=-10.0, high=10.0, size=(observations,variabels))
axes_z = np.random.uniform(low=-10.0, high=10.0, size=(observations,variabels))
input_data = np.column_stack((axes_x,axes_z))

noise = np.random.uniform(low=-0.1, high=0.1, size=(observations,variabels))
target_data = 2 * axes_x + 3 * axes_z + 5 + noise

# salvar os dados em formato de tensor
np.savez('tf_data', inputs=input_data, targets=target_data)
```

#### Visualizar os dados

```python
target_data = target_data.reshape(observations,)
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(axes_x, axes_z, target_data)
plt.show()
```

## Rede Neural com uma camada

### Fun√ß√µes de Erro:

- Avalia o qu√£o distante estamos dos dados reais em modelos supervisionados. As fun√ß√µes mais comuns s√£o:
    - **L2 Norm** para modelos de regress√£o: $\sum_{i}(\mbox{output}_{i}-\mbox{target}_{i})^{2}$
    - **Cross Entropy** para modelos de classifica√ß√£o: $-\sum_{i}(\mbox{target}_{i}*\ln(\mbox{output}_{i}))$
    * Esse √∫ltimo compara a probabilidade de classifica√ß√£o para cada poss√≠vel categoria

### Fun√ß√£o de Ativa√ß√£o

- Essa √© a fun√ß√£o que estar√° encarregada de otimizar os pesos da nossa rede neural:
    - **Gradiente Descendente**: nessa otimiza√ß√£o a ideia √© atingir o m√≠nimo global da fun√ß√£o de erro  da forma mais otimizada poss√≠vel: $w_{i+1} = w_{i} - \eta * f'(w_{i}) $
    - Onde $\eta$ √© a taxa de aprendizado da fun√ß√£o. Observe que esse par√¢metro deve ser escolhido com cautela porque n√£o queremos um valor t√£o pequeno que a converg√™ncia do resultado da rede demore muito para ocorrer e t√£o pouco queremos um valor alto que ultrapasse o local do m√≠nimo global da fun√ß√£o de erro. 

```python
training_data = np.load('tf_data.npz')
```

```python
input_size = 2
output_size = 1

# inicia modelo
model = tf.keras.Sequential(
                [tf.keras.layers.Dense(output_size,
                                      kernel_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1),
                                      bias_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1))
                ])

# inclui fun√ß√£o de otimiza√ß√£o e fun√ß√£o objetivo
custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=custom_optimizer, loss='mean_squared_error')

# fit modelo
# verbose=0 -> n√£o printa o resultado 
# verbose=1 printa do resultado de cada √©poca
# verbose=2 printa resultados mais condensados
model.fit(training_data['inputs'], training_data['targets'], epochs=10, verbose=2)
```

```python
# avalia os valores do pesos e do vies
weights = model.layers[0].get_weights()[0]
biases = model.layers[0].get_weights()[1]
print('Pesos:',  weights)
print('Vi√©s:', biases)
```

```python
# previs√£o de novos valores
output = model.predict_on_batch(training_data['inputs'])
print(np.sum((output-target_data)**2)/len(output))
```
